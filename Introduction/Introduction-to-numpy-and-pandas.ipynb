### Some linear Algebra

```python
from numpy import linalg as LA
A = np.arange(9).reshape(3, 3)
A
```

```python
print(np.linalg.det(A))
print(LA.matrix_rank(A))
```

```python
w, v = LA.eig(A)
w
```

```python
A = A + np.eye(3, 3)
print(LA.matrix_rank(A))   # Full rank, hence invertible
```

```python
b = np.array([1, 2, 5])
x = np.linalg.solve(A, b)  # A^{-1}b

np.matmul(A, x)            # Should give back b
```
